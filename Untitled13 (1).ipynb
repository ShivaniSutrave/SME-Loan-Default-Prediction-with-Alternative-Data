```python
import os
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
from google import genai

os.environ["GEMINI_API_KEY"] = "your_key_here"

startup_knowledge = [
"TOPIC: How to get your first 100 customers. Do things that don't scale. Go to customers manually. Start with your personal network first. Post in Reddit, LinkedIn, Facebook groups. Offer it free to 10 users, get feedback, charge the next 10.",
"TOPIC: When to raise funding. Best time to raise is when you don't need it. Raise when you have 10 paying customers and 10% monthly growth. Start fundraising when you have 6 months runway left.",
"TOPIC: Product Market Fit. Signs you have PMF: users come back without asking, word of mouth grows naturally. Sean Ellis test: 40% of users say very disappointed if product disappeared.",
"TOPIC: How to price your SaaS product. Most founders price too low. SMB customers: $29 to $199 per month. Mid market: $500 to $2000 per month. Always offer annual pricing with 2 months free.",
"TOPIC: Managing burn rate. Runway equals cash in bank divided by monthly burn rate. Always maintain minimum 12 months runway. Cut software subscriptions first before touching team.",
"TOPIC: How to hire first employees. First 10 hires define your culture. Hire slow fire fast. Employee equity 0.5% to 2% with 4 year vesting. Hire generalists early.",
"TOPIC: Growth strategies. Focus on one growth channel deeply. Fix retention first. B2B SaaS: LinkedIn outbound and cold email work best. LTV to CAC must be above 3.",
]

splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=20)
all_chunks = []
for doc in startup_knowledge:
    all_chunks.extend(splitter.split_text(doc))
print(f"✅ Split into {len(all_chunks)} chunks")

print("⏳ Loading embedding model... wait 2 minutes")
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

vectorstore = Chroma.from_texts(
    texts=all_chunks,
    embedding=embeddings,
    persist_directory='/content/drive/MyDrive/AI_Business_Advisor/data/vectorstore'
)
print(f"✅ Vector database ready! Total vectors: {len(all_chunks)}")
```

```python
app_code = '''
import streamlit as st
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
from google import genai

st.set_page_config(page_title="AI Startup Advisor", page_icon="🚀", layout="wide")

st.markdown("""
<style>
h1 { color: #00d4ff; }
</style>
""", unsafe_allow_html=True)

st.title("🚀 AI Startup Business Advisor")
st.markdown("*Powered by RAG + Gemini AI*")
st.markdown("---")

with st.sidebar:
    st.header("Settings")
    api_key = st.text_input("Gemini API Key", type="password", placeholder="AIzaSy...")
    st.markdown("---")
    st.markdown("### Try These Questions")
    for q in [
        "How do I get my first 100 customers?",
        "When should I raise funding?",
        "How to find product market fit?",
        "My burn rate is too high, help!",
        "How should I price my SaaS?",
        "When should I hire first employee?",
    ]:
        if st.button(q, use_container_width=True):
            st.session_state.clicked_question = q

if "messages" not in st.session_state:
    st.session_state.messages = []
    st.session_state.messages.append({
        "role": "assistant",
        "content": "Hello! I am your AI Startup Advisor. Ask me anything about fundraising, growth, hiring, pricing, or product market fit!"
    })

for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.markdown(message["content"])

def get_answer(question, api_key):
    embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
    vectorstore = Chroma(persist_directory="vectorstore", embedding_function=embeddings)
    docs = vectorstore.similarity_search(question, k=3)
    context = "\\n".join([d.page_content for d in docs])
    prompt = f"""You are an expert startup advisor with 20 years experience.
Use context below to give specific actionable advice.
End with 2 concrete next steps the founder can take TODAY.
Context: {context}
Question: {question}
Expert Advice:"""
    client = genai.Client(api_key=api_key)
    response = client.models.generate_content(model="gemini-2.5-flash", contents=prompt)
    return response.text

if "clicked_question" in st.session_state:
    question = st.session_state.pop("clicked_question")
    st.session_state.messages.append({"role": "user", "content": question})
    with st.chat_message("user"):
        st.markdown(question)
    with st.chat_message("assistant"):
        with st.spinner("Thinking..."):
            answer = get_answer(question, api_key)
            st.markdown(answer)
            st.session_state.messages.append({"role": "assistant", "content": answer})

if prompt := st.chat_input("Ask your startup question..."):
    st.session_state.messages.append({"role": "user", "content": prompt})
    with st.chat_message("user"):
        st.markdown(prompt)
    with st.chat_message("assistant"):
        with st.spinner("Thinking..."):
            if not api_key:
                st.error("AIzaSyCTs65o_P3IxdYoBmeVChUQRE97bpBOu0A")
            else:
                answer = get_answer(prompt, api_key)
                st.markdown(answer)
                st.session_state.messages.append({"role": "assistant", "content": answer})
'''

with open('/content/drive/MyDrive/AI_Business_Advisor/app/advisor.py', 'w') as f:
    f.write(app_code)

print("✅ App file saved!")
```
